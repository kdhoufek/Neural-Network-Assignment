A. Build a baseline model

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.

```python
#First we import all necessary libraries for this project.
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Import our dataset
df = pd.read_csv(r"C:\Users\kdhou\Downloads\concrete_data.csv")
df.head()
```

```python
#Split data into features and target variables
features = df[df.columns[:-1]]
target = df[df.columns[-1]]
n_features = features.shape[1]
n_features
```

```python
#Define the baseline model as a function
def baseline_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_features,)))
    model.add(Dense(1))
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return(model)
```

1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the 
train_test_split helper function from Scikit-learn.

```python
#Split data into train-test sets
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = None)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

2. Train the model on the training data using 50 epochs.

```python
#Call the model
model = baseline_model()

#Fit the model as per above
model.fit(x_train, y_train, epochs = 50, verbose = 1)
```

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

```python
#Call the evaluate method on the testing data
score = model.evaluate(x_test, y_test, verbose = 0)
print('Mean squared error of the model:', score)
```

4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

5. Report the mean and the standard deviation of the mean squared errors.

```python
#Create initial empty list
mean_squared_errors = []

#Loop through 50 iterations and append results to list
i = 1
while i <= 50:
    #Create new train/test split each iteration
    x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = None)

    #Call model function on new splits
    model = baseline_model()
    model.fit(x_train, y_train, epochs = 50, verbose = 0)

    #Evaluate on test data
    mse = model.evaluate(x_test, y_test, verbose = 0)

    #Append to list and increment i
    mean_squared_errors.append(mse)
    i += 1

#Calculate mean and stdev of MSE's
mean = np.average(mean_squared_errors)
stdev = np.std(mean_squared_errors)
print("Mean of the MSE's is: {} \n Standard Deviation of the MSE's is: {}".format(mean, stdev))
```

We see that the average MSE of the output is approximately 366.1, and the standard deviation of these MSE's is approximately 432.0. 

Let's see if we can reduce this error.

B. Normalize the data (5 marks)

Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

How does the mean of the mean squared errors compare to that from Step A?

```python
#Normalize by subtracting mean and dividing by standard deviation
#Note that we could normalize the target variable, but it is not necessary.
features_normalized = (features - features.mean())/features.std()
features_normalized.head()
```

```python
#Pass the normalized data through the model and find the mean and stdev of the MSE's as before
#Create initial empty list
mean_squared_errors = []

#Loop through 50 iterations and append results to list
i = 1
while i <= 50:
    #Create new train/test split each iteration, this time using normalized features data
    x_train, x_test, y_train, y_test = train_test_split(features_normalized, target, test_size = 0.3, random_state = None)

    #Call model function on new splits
    model = baseline_model()
    model.fit(x_train, y_train, epochs = 50, verbose = 0)

    #Evaluate on test data
    mse = model.evaluate(x_test, y_test, verbose = 0)

    #Append to list and increment i
    mean_squared_errors.append(mse)
    i += 1

#Calculate mean and stdev of MSE's
mean = np.average(mean_squared_errors)
stdev = np.std(mean_squared_errors)
print("Mean of the MSE's is: {} \n Standard Deviation of the MSE's is: {}".format(mean, stdev))
```

Our average MSE for normalized data is approximately 340.1. Compared to an average of 366.1 with the unnormalized data, this is better but not by much.

However, our standard deviation went from 432.0 down to only 98.6, a considerable decrease. So we now have much less variation in our error term, which is good.

C. Increate the number of epochs (5 marks)

Repeat Part B but use 100 epochs this time for training.

How does the mean of the mean squared errors compare to that from Step B?

```python
#We simply repeat the same code on the normalized data as in part B, but using epochs = 100 on the model.fit() method.
#Create initial empty list
mean_squared_errors = []

#Loop through 50 iterations and append results to list
i = 1
while i <= 50:
    #Create new train/test split each iteration, this time using normalized features data
    x_train, x_test, y_train, y_test = train_test_split(features_normalized, target, test_size = 0.3, random_state = None)

    #Call model function on new splits
    model = baseline_model()
    model.fit(x_train, y_train, epochs = 100, verbose = 0)

    #Evaluate on test data
    mse = model.evaluate(x_test, y_test, verbose = 0)

    #Append to list and increment i
    mean_squared_errors.append(mse)
    i += 1

#Calculate mean and stdev of MSE's
mean = np.average(mean_squared_errors)
stdev = np.std(mean_squared_errors)
print("Mean of the MSE's is: {} \nStandard Deviation of the MSE's is: {}".format(mean, stdev))
```

Our Mean of the MSE's has now dropped from 340.1 down to 165.4, which is huge. Our standard deviation also dropped from 98.6 down to 20.3.

We can conclude that increasing the iterations (epochs) to 100 has improved the model significantly.

Repeat part B but use a neural network with the following instead:

- Three hidden layers, each of 10 nodes and ReLU activation function.

How does the mean of the mean squared errors compare to that from Step B?

```python
#We first define the new model as three_layer_model
def three_layer_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_features,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return(model)
```

```python
#Apply this model using the code from step C to get our final output
#Create initial empty list
mean_squared_errors = []

#Loop through 50 iterations and append results to list
i = 1
while i <= 50:
    #Create new train/test split each iteration, this time using normalized features data
    x_train, x_test, y_train, y_test = train_test_split(features_normalized, target, test_size = 0.3, random_state = None)

    #Call model function on new splits
    model = three_layer_model()
    model.fit(x_train, y_train, epochs = 100, verbose = 0)

    #Evaluate on test data
    mse = model.evaluate(x_test, y_test, verbose = 0)

    #Append to list and increment i
    mean_squared_errors.append(mse)
    i += 1

#Calculate mean and stdev of MSE's
mean = np.average(mean_squared_errors)
stdev = np.std(mean_squared_errors)
print("Mean of the MSE's is: {} \nStandard Deviation of the MSE's is: {}".format(mean, stdev))
```

Final Mean of MSE's is 89.16, which is by far the lowest error we have seen. Standard deviation is 21.7 which is comparable to our output in part C.

Overall we can conclude that adding hidden layers has greatly improved the model by reducing the error term.
